# Export audio_sample.rs

Collects test clips from the dataset and writes `src/audio_sample.rs` for the Rust runner.

```python
%load_ext autoreload
%autoreload 2
import sys
from pathlib import Path
```

```python
repo_root = Path.cwd().parent
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))
```

```python

from building_tensorflow.utils import (
    SAMPLE_RATE,
    TARGET_AUDIO_LEN_TIME,
    DATASET_ROOT,
    SRC_DIR,
    collect_test_clips_for_rs,
    write_audio_sample_rs,
)

OUT_AUDIO_RS = SRC_DIR / "audio_sample.rs"
```

```python
clips = collect_test_clips_for_rs(
    DATASET_ROOT / "testing",
    sample_rate=SAMPLE_RATE,
    target_len=TARGET_AUDIO_LEN_TIME,
    num_per_label=2,
)

write_audio_sample_rs(
    OUT_AUDIO_RS,
    clips,
    SAMPLE_RATE,
    generator_name="building_tensorflow/export_audio_sample_rs.ipynb",
)

print("Wrote", OUT_AUDIO_RS, "clips=", len(clips), "samples_per_clip=", len(clips[0][1]))
```
